Como pode ser verificado muitas colunas estão com o tipo inadequado. Vamos ter que tratar essas colunas.
 

Colocando o tipo correto na colunas que indicam tempo, vou volocar a data e a hora em uma mesma coluna

```python

# Definindo o local para pt_BR, para fazer a conversao do nome dos meses nos números correspondentes aos meses.
locale.setlocale(locale.LC_TIME, 'pt_BR.UTF-8')
#lista para guardar o número dos meses
list_number_month = list()
#lista para guardar o dia das partidas
list_number_of_day = list()
#lista para guardar o dia da semana
list_day_of_week = list()
#lista para guardar o horario do jogo
list_hour = list()
for data in df_flamengo['data_jogo']:
    list_number_month.append(datetime.strptime(data.split()[1],'%B').month)
    list_number_of_day.append(int(data.split()[2]))
    list_day_of_week.append(data.split()[0])
for hour in df_flamengo['horario_jogo']:
    list_hour.append(hour.split()[0])

#remonemando a coluna data_jogo para data_horario_jogo
df_flamengo = df_flamengo.rename({'data_jogo':'data_horario_jogo'})
for i in range(0,len(list_number_month)):
    year = 2023
    month = list_number_month[i]
    day = list_number_of_day[i]
    hour = int(list_hour[i].split(':')[0])
    minutes = int(list_hour[i].split(':')[1])
    seconds = 0
    df_flamengo.loc[i,'data_horario_jogo'] = datetime(year,month,day,hour,minutes,0).strftime(f'%d/%m/%Y %H:%M:%S')

#mudando a posição da coluna
df_flamengo.insert(10,'data_horario_jogo',df_flamengo.pop('data_horario_jogo'))
#removendo as colunas data_jogo e horario_jogo
df_flamengo = df_flamengo.drop(columns=['data_jogo','horario_jogo'])
```

Agora vamos tratar algumas colunas que possui algumas palavras redundantes, por exemplo, na coluna tecnico_time_casa o nome do técnico é escrito assim Técnico: Nome do técnico, como a coluna ja especifica o que esse nome significa vamos deixar somente o nome. 

```python
columns  = ['tecnico_time_casa','capitao_time_casa','tecnico_time_visitante','capitao_time_visitante','estadio_jogo','arbitro_jogo',
            'publico_jogo','renda_jogo']
#função que remove a redundancia de algumas colunas
def remove_redundancy(columns,df_flamengo):
    for name_colum in columns:
        for i in range(0,38):
            df_flamengo.loc[i,f'{name_colum}'] = df_flamengo.loc[i,f'{name_colum}'].split(':')[1].lstrip()
remove_redundancy(columns,df_flamengo)
```
